```python
import pandas, numpy, scipy.stats
```

```python
test = pandas.DataFrame()
test["TF"] = list("ABCABABC")
test["Method"] = list("11122333")
test["AUC"] = scipy.stats.norm.rvs(0.5, 0.1, len(test))
# test.set_index(["TF", "Method"], inplace=True)
test
```

```python
# just to see what this looks like:
grouped = test.groupby(["TF", "Method"]).mean()
grouped
```

```python
pivoted = test.pivot("TF", "Method", "AUC")
pivoted
```

```python
# you can do the subtractions in a smarter way:
pivoted["1-2"] = pivoted["1"] - pivoted["2"]
pivoted["1-3"] = pivoted["1"] - pivoted["3"]
pivoted["2-3"] = pivoted["2"] - pivoted["3"]
pivoted
```

```python
unstacked = pivoted.unstack()
finished = pandas.DataFrame(unstacked, columns=["AUC"]).reset_index()
finished
```
