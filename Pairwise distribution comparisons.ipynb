```python
import scipy.stats, numpy, scipy, seaborn, pylab, pandas
seaborn.set_style("white")
%matplotlib inline
```

```python
N = 1000
# true = scipy.stats.uniform.rvs(-5, 10, size=N)
true = scipy.stats.norm.rvs(0, 2, size=N)
x = true + scipy.stats.norm.rvs(0, 0.5, size=N)
y = true + scipy.stats.norm.rvs(0, 0.5, size=N)
```

```python
pylab.plot(x, y, 'bo', alpha=0.5)
seaborn.despine(offset=10)
pylab.show()
```

```python
bins = numpy.arange(-8,8,0.1)
args = {"bins": bins, "histtype": "step", "lw": 2}
pylab.hist(x, **args)
pylab.hist(y, **args)
seaborn.despine(offset=10)
pylab.show()
```

```python
seaborn.jointplot(x, y, kind="kde")
# seaborn.despine(offset=10)
pylab.show()
```

```python
seaborn.jointplot(x, y, kind="hex")
# seaborn.despine(offset=10)
pylab.show()
```

```python
seaborn.jointplot(x, y, kind="resid")
# seaborn.despine(offset=10)
pylab.show()
```

```python
seaborn.jointplot(x, y, kind="reg")
# seaborn.despine(offset=10)
pylab.show()
```

```python
seaborn.violinplot([x, y])
seaborn.despine(offset=10)
pylab.show()
```

```python
seaborn.violinplot([x, y], join_rm=True)
seaborn.despine(offset=10)
pylab.show()
```

```python
seaborn.boxplot([x, y])
seaborn.despine(offset=10)
pylab.show()
```

```python
seaborn.boxplot([x, y], join_rm=True)
seaborn.despine(offset=10)
pylab.show()
```

```python
df = pandas.DataFrame({"x":x, "y":y})
pylab.hist(df[df["x"].abs() > 0.5]["x"].values, **args)
pylab.hist(df[df["x"].abs() > 0.5]["y"].values, **args)
pylab.show()
```
